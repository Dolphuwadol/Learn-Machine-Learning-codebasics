### Gradient Descent

```python
import numpy as np
```

```python
def gradient_descent(x,y):
    m_curr = b_curr = 0 # ค่าเริ่มต้นของค่าความชัน และ ค่าตัดเส้น
    iterations = 100 #กำหนดรอบในการวนซ้ำ
    n = len(x) 
    learning_rate = 0.005
    for i in range(iterations): # วนลูปเพื่อคำนวนค่าความคลาดเคลื่อนของการทำนายและค่า cost
        y_predicted = m_curr * x + b_curr
        cost = (1/n) * sum([val**2 for val in (y-y_predicted)]) # cost funtion หรือ loss funtion หรือ mse
        md = -(2/n) * sum(x*(y-y_predicted)) #dif ค่าความชัน
        bd = -(2/n) * sum(y-y_predicted) # dif ค่าตัดเส้น
        m_curr = m_curr - learning_rate * md # อัปเดตค่าความชัน
        b_curr = b_curr - learning_rate * bd # อัปเดตค่าตัดเส้น
        print("m {}, b {}, cost {} iteration {}".format(m_curr, b_curr,cost,i))
        

x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])

gradient_descent(x,y)
#ต้องการเข้าใกล้ m 2, b 3 , และ cost ต่ำที่สุด
# หาค่า intercept กับ coef  
```
