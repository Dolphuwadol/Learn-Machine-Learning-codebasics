### Solve a problem to predict hand written digit

```python
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
import numpy as np 
import pandas as pd 
```

```python
digits = load_digits()
```

```python
dir(digits)
```

```python
df = pd.DataFrame(data=digits['data'], columns=digits['feature_names'])
df['target'] = digits['target']
df['target_names'] = digits['target_names'][df['target']]
df
```

```python
digits.data[0] #รูปภาพ 8*8 
```

```python
plt.gray()
for i in range(5):
    plt.matshow(digits.images[i])
```

```python
digits.target[0:20]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target,
                                                   train_size = 0.8)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test, y_test)
```

```python
model.predict(X_test)
```

```python
plt.matshow(digits.images[650]) #แสดงรูปที่ 650
```

```python
digits.target[650] #คำตอบของรูป 650
```

```python
model.predict(digits.data[[650]]) #model ทำนายที่รูป 650
```

### Confusion Matrix

```python
y_predicted = model.predict(X_test)
from sklearn.metrics import confusion_matrix 
cm = confusion_matrix(y_test, y_predicted)
```

```python
import seaborn as sn
plt.figure(figsize = (10, 7))
sn.heatmap(cm, annot= True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
# ex. รูปที่ป้อนเข้าไปเป็นเลข 2 แต่ model ทำนายเป็นเลข 1 
```

```python
from sklearn import datasets
```

```python
iris = datasets.load_iris()
dir(iris)
```

```python
df = pd.DataFrame(data=iris['data'], columns=iris['feature_names'])
df['target'] = iris['target']
df['target_names'] = iris['target_names'][df['target']]
df
```

```python
iris.data[0:5]
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target,
                                                   train_size = 0.8)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
X_test
```
